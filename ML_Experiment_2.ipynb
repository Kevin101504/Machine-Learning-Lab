```python
df_ffill = df.copy()
df_ffill['FUELCONSUMPTION_CITY'] = df_ffill['FUELCONSUMPTION_CITY'].fillna(method='ffill')
print(df_ffill.isnull().sum())
```

```python
df_bfill = df.copy()
df_bfill['FUELCONSUMPTION_CITY'] = df_bfill['FUELCONSUMPTION_CITY'].fillna(method='bfill')
print(df_bfill.isnull().sum())
```

```python
# Text to Numerical:
#   categorical
#   one hot encoder
#   label encoder(used in places including classes)
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
le = LabelEncoder()
df_encoded['VEHICLECLASS'] = le.fit_transform(df_encoded['VEHICLECLASS'])
print(df_encoded['VEHICLECLASS'])
```

```python
from sklearn.preprocessing import OneHotEncoder
df_encoded = df.copy()
ohe = OneHotEncoder()
vehicleclass_encoded = ohe.fit_transform(df_encoded[['VEHICLECLASS']]).toarray()
print(vehicleclass_encoded)
```
